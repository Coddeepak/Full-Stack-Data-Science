```python
def greater(n1,n2):

    if n1>n2:
        return(n1)
    else:
        return(n2)
greater(100,200)
```

```python
# format:
# lambda <arg1>,<arg2>: <if_output> <if_con> else <else_op>
greater = lambda a,b:a if a>b else b
greater(200,305)
```
